```python
import pandas as pd
from pyproj import Proj
import folium
from folium.plugins import HeatMap
```

```python
# read data
df = pd.read_csv("intersects_190710.csv")
```

```python
converter = Proj(proj='utm', zone=54, ellps='WGS84')
```

```python
def CONVERT(df):
    lon, lat = converter(df['x'], df['y'], inverse=True)
    return [lon, lat]
```

```python
tmp = df.apply(CONVERT, axis=1)
tmp
```

```python
mean_lon, mean_lat = converter(df['x'].mean(),df['y'].mean(),inverse=True)
```

```python
# map
map = folium.Map(location=[mean_lat,mean_lon])
```

```python
for row in tmp:
    folium.Marker(location = [row[1],row[0]]).add_to(map)
```

```python
map
```

```python
map.save('test.html')
```

```python
data = list()
for row in tmp:
    location = [row[1],row[0]]
    data.append(location)
    
data
```

```python
hmap = folium.Map(location=[mean_lat,mean_lon])
HeatMap(data, radius=15).add_to(hmap)
hmap.save('Heatmap.html')
```
